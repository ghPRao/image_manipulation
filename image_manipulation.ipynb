```python
chelsea = data.chelsea()
```

```python
plt.figure(figsize=(8, 8))
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(chelsea,  interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
chelsea[:10] = 10
mask = chelsea < 100
chelsea[mask] = 255
```

```python
plt.figure(figsize=(8,8))
plt.imshow(chelsea,  interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
inds_x = np.arange(len(chelsea))
inds_y = (4 * inds_x) % len(chelsea)
chelsea[inds_x, inds_y] = 0

l_x, l_y = chelsea.shape[0], chelsea.shape[1]
X, Y = np.ogrid[:l_x, :l_y]
outer_disk_mask = (X - l_x / 2)**2 + (Y - l_y / 2)**2 > (l_x / 2)**2
chelsea[outer_disk_mask] = 0
```

```python
plt.figure(figsize=(8,8))
plt.imshow(chelsea, interpolation='nearest')
plt.axis('off')
plt.show()
```
